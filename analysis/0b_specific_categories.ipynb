```python
import django, sys, os
sys.path.append('/home/max/software/django-tmv/tmv_mcc-apsis/BasicBrowser')
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "BasicBrowser.settings")
django.setup()

from scoping.models import *

import pandas as pd

```

```python
df = pd.read_csv('../data/0_labelled_documents.csv')
```

```python
pred_cats = [c for c  in df.columns if "12 - " in c ]
pred_cats
```

```python
catids = [710,320,321,323,322,324]
cats = [Category.objects.get(pk=cid) for cid in catids]
```

```python
data = []

for c in cats:
    cname = f"{c.level} - {c.name}"

    for ec in c.equivalents.all().order_by('name','level'):
        ecname = f"{ec.level} - {ec.name}"

        if "hidden" in ecname:
            continue
        if ecname in df.columns:
            if ec.level==4:
                data.append({
                    "Broad category": cname.split(" - ")[1],
                    "Specific category": ecname.split(" - ")[1],
                    "Labelled documents": df[ecname].sum()
                })
                
cat_df = pd.DataFrame.from_dict(data)
cat_df.to_csv('../data/category_label_numbers.csv',index=False)
cat_df.head()
```
